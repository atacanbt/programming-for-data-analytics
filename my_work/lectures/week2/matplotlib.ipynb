## import the packages

matplotlib and numpy

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Some data

generate a data

np.arrange() will create a series of numbers

```python
x = 0.5 + np.arange(8)
y = [1,3,5,8,3,6,4,6]
y2 = [3,5.6,9.4,9.1,4.6,7.7,8.1,9]
```

### Making a plot

using

 **`fig, ax = plt.subplots()`**

and

**`plt.show()`**

make a plot, eg:bar

```python
fig, ax = plt.subplots()
ax.bar(x,y)
ax.plot(x,y2)
#ax.scatter(x,y)
#ax.fill_between(x,y,y2)
#ax.stackplot(x,y,y2)
#ax.stairs(y)
#ax.stem(x,y)

plt.show()
```

### set()

setting paramaters on plot using. [supported properties](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set.html)

**`set()`**

it can take paramaters like `xlim` `xlabel` `title`

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set(xlabel="x axis label", ylabel="y axis label", title="Title")

plt.show()
```

### The axis funtion

Or we can use the functions in axis individually 

[`set_title`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.htmlhttps://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html) [`set_xlabel`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html) [`set_ylabel`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html)

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("Title")
ax.set_xlabel("x axis label")
ax.set_ylabel("y axis label")

plt.show()
```

### Format the labels

Use a formatter function to format the labels using

[`ax.xaxis.set_major_formatter(formatter_fun)`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axis.Axis.set_major_formatter.html)

#### The function that will do the formatting

It takes two paramaters the value on the axis and the position of the tick. It should return text as if should appear

```python
def currency_formatter(x, pos):
    s = f'${x}M'
    return s
```

#### Pass the function to

`ax.xaxis.set_major_formatter(formatter_fun)`

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("Title")
ax.set_xlabel("x axis label")
ax.set_ylabel("y axis label")

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

plt.show()
```

### Add text

Using [`text()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.text.html)

`ax.text(x,y,"text to add", fontsize=10, verticalalignment="center")`

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("Title")
ax.set_xlabel("x axis label")
ax.set_ylabel("y axis label")

ax.text(0,8,"blahh", fontsize=10, verticalalignment="center")

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

plt.show()
```

### Annotations

[`ax.annotate`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html)("text", (x_pointer, y_pointer), xytext=(x,y), [`arrowprops=dict(facecolor='color', shrink=0.05`]((https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.arrow.html)))

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("Title")
ax.set_xlabel("x axis label")
ax.set_ylabel("y axis label")

#ax.text(0,8,"blahh", fontsize=10, verticalalignment="center")
ax.annotate("max", (3.2, 8), xytext=(1,9), arrowprops=dict(facecolor='black', shrink=0.05))

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

plt.show()
```
